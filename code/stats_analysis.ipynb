```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'..\..\..\..\data\census\98-401-X2016066_English_CSV_data.csv')
```

```python
#Filter 'DIM: Profile of Census Subdivisions (2247)' for “Total - Knowledge of official languages for the total population excluding institutional residents - 100% data”
df_rows_filtered = df.loc[df["DIM: Profile of Census Subdivisions (2247)"] == "Total - Knowledge of official languages for the total population excluding institutional residents - 100% data"]
df_rows_filtered
```

```python
# filtering the resulting DataFrame for specific columns
df_cols_filtered = df_rows_filtered[['GEO_CODE (POR)', 'Dim: Sex (3): Member ID: [1]: Total - Sex']]
df_cols_filtered
```

```python
# Renaming the columns
df_renamed = df_cols_filtered.rename(
    columns = {
        "GEO_CODE (POR)" : "GEOCODE",
        "Dim: Sex (3): Member ID: [1]: Total - Sex" : "Total",
    }
)

df_renamed
```

```python
# Make GEOCODE the index
df_index = df_renamed.set_index('GEOCODE')
df_index
```

```python
#Filter out rows where the Median Income column is “x”, “F”, or “..”
condition_1 = df_index['Total'] != 'x'
condition_2 = df_index['Total'] != 'F'
condition_3 = df_index['Total'] != '..'
df_filtered_total = df_index[condition_1 & condition_2 & condition_3]
df_filtered_total
```

```python
#Filter 'DIM: Profile of Census Subdivisions (2247)' for “Total - Knowledge of official languages for the total population excluding institutional residents - 100% data”
df_rows_filtered_bilingual = df.loc[df["DIM: Profile of Census Subdivisions (2247)"] == "English and French"]
df_rows_filtered_bilingual
```

```python
# filtering the resulting DataFrame for specific columns
df_cols_filtered_bilingual = df_rows_filtered_bilingual[['GEO_CODE (POR)', 'Dim: Sex (3): Member ID: [1]: Total - Sex']]
df_cols_filtered_bilingual
```

```python
# Renaming the columns
df_renamed_bilingual = df_cols_filtered_bilingual.rename(
    columns = {
        "GEO_CODE (POR)" : "GEOCODE",
        "Dim: Sex (3): Member ID: [1]: Total - Sex" : "French and English",
    }
)

df_renamed_bilingual
```

```python
# Dropping duplicates
df_unique_bilingual = df_renamed_bilingual.drop_duplicates(subset = 'GEOCODE')
df_unique_bilingual
```

```python
# Make GEOCODE the index
df_index_bilingual = df_unique_bilingual.set_index('GEOCODE')
df_index_bilingual
```

```python
# Joining both data frames
table_languages_canada = pd.merge(df_filtered_total, df_index_bilingual, how = 'inner', on = 'GEOCODE')
table_languages_canada
```

```python
# Calculating the percentage of bilingual speakers
table_languages_canada['Percentage Bilingual Speakers'] = (table_languages_canada['French and English'].astype(np.float64)/table_languages_canada['Total'].astype(np.float64))*100
table_languages_canada
```

```python
# summary statistics using describe method
table_languages_canada['Percentage Bilingual Speakers'].describe()
```

```python
#Selective descriptive statistics
agg_result = table_languages_canada.agg(
    {
        'Percentage Bilingual Speakers': ['min','max', 'mean', 'median', 'std']
    }
)

agg_result_string = agg_result.to_string()

with open('summary_stats.txt', 'w') as outfile:
    outfile.write(agg_result_string)
```
